###Stemming- reducing a word to its root word

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word,"----------->",stemming.stem(word))  ##not proper root word for few
```

```python
stemming.stem("agriculture")
```

###RegexpStemmer -- > removes the string matching the Regex and gives the root word

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4) #it will stem only if the root len is 4
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('acids')
```

```python
reg_stemmer.stem('ingeating')
```

##### Snowball Stemmer ->Porter2 algorithm which is more accurate than the porterstemmer


```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"-------->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),snowballstemmer.stem("fairly")
```

```python
stemming.stem("goes"),snowballstemmer.stem("goes") ## some stems are still incorrect
```
