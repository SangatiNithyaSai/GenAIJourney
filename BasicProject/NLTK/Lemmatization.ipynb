### Lemmatizer - it provides root word rather than the **root stem by Stemmer module. 
### WordNet Lemmatizer - it is a wrapper class around wordnet corpus. 

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
''' 
we need to provide the parts of speech
Noun-n
verb-v
adjective- a
adverb-r
'''
lemmatizer.lemmatize('going',pos='v')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized","dating"]
```

```python
for word in words:
    print(word+"------------>"+lemmatizer.lemmatize(word,pos='v'))
```

```python
#Far better than stemming, but need to provide POS
lemmatizer.lemmatize("goes",pos='v'),lemmatizer.lemmatize("fairly",pos='v'),lemmatizer.lemmatize("sportly",pos='v')
```
