###  Named Entity Recognition
It identifies a word as a Person,Place or Date.
<br>
Example:
<br>
Person - Nithya Sai
<br>
Place - Hyderabad
<br>
Money- 2 lakhs
<br>
Time - 4.44pm


```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
words
```

```python
tag_elements=nltk.pos_tag(words)
tag_elements
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
word_NE=nltk.ne_chunk(tag_elements)
word_NE
#nltk.ne_chunk(tag_elements).draw() # need to install Tkinter to see the visualization
```
