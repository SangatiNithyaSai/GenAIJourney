###Tokenization with NLTK


```python
corpus="""Hi all, This is a simulation work.
we are going to learn about the tokenization.
"""
```

```python
print(corpus)
```

```python
#converts each senences as tokens
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
```

```python
documents
```

```python
##  amking every work as token
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus) # . , also considered as tokens
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
# to seperate 's 
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize("My name's is Nithya Sai.")
```

```python
print(word_tokenize("My name's is Nithya Sai."))
print(wordpunct_tokenize("My name's is Nithya Sai."))
```

```python
## to consider . as part of the word other than the last .
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
