```python
# infinite generator

def infinite_chai():
    count=1
    while True:
        yield f"Refill #{count}"
        count+=1
refill = infinite_chai()
user2 = infinite_chai()

for _ in range(5):
    print(next(refill))

print("\n")
##value does not change
for _ in range(5):
    print(next(user2))

```

```python
#send generator

def chai_customer():
    print("Welcome! what chai do you like?")
    order=yield
    while True:
        print(f"Preparing {order}")
        order=yield

stall=chai_customer()
next(stall)
stall.send("Masala Chai")
```

```python
# let me send another value
stall.send("Ginger Chai")
```

```python
#lets close
stall.close()
```

```python
stall.send("allam") # we have closd the generator
```

```python
def local_chai():
    yield "Masala Chai"
    yield "Ginger Chai"
def imported_chai():
    yield "Niloufer"
    yield "Oolong"

def full_menu():
    yield from local_chai()
    yield from imported_chai()

for chai in full_menu():  ## internally calls next
    print(chai)
```

```python
def chai_stall():
    try:
      while True:
        order=yield "waiting for chai order"
    except:
       print("Stall Closed,No more chai")
```

```python
stall=chai_stall()
print(stall)
next(stall)
# stall.send("masala chai")
```
