```python
#decorator
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")

    return wrapper

@my_decorator
def greet():
    print("Hello! Inside the Greet function")

greet()
```

```python
print(greet.__name__)
```

```python
# to send the function details we need to use the wrap functionality
#decorator
from functools import wraps
def my_decorator(func):
    @wraps(func)
    def wrapper():
        print("Before function")
        func()
        print("After function")

    return wrapper

@my_decorator
def greet():
    print("Hello! Inside the Greet function")

greet()
print(f"\n\n{greet.__name__} ")
```

```python
# argumenet passing

def arg_decorator(func):
    @wraps(func)
    def wrapper(*args,**kwargs):
        print(f"Calling {func.__name__}")
        result=func(*args,**kwargs)
        print(f"Finished Calling {func.__name__}")
        print(result) #not required
    return wrapper

@arg_decorator
def chai_maker(type,sugar="No"):
    print(f"{type} chai with {sugar} sugar")

chai_maker("Ilachi","2 spoons")
```

```python
# user access
def verify_user(func):
    @wraps(func)
    def wrapper(user_role):
        if user_role !="admin":
            print("Access Denied for ",user_role)
        else:
            func(user_role)
    return wrapper

@verify_user
def valid_user(role):
    print(f"Chai for {role} is getting ready")

valid_user("admin")
valid_user("engineer")
```
