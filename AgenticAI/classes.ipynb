```python
## Init method
class Chai:
    def __init__(self,type_,size):
        self.type=type_
        self.size=size
    def summary(self):
        return f"{self.type} of {self.size} ml"
    
chai1=Chai("Masala",200)
print("Chai1:")
print(chai1.summary())
chai2=Chai("Ginger",100)
print("Chai2:")
print(chai2.summary())
```

```python
#Inheritence

class BaseChai:
    def __init__(self,type_):
        self.type=type_
    def prepare(self):
        print(f"Preparing {self.type} Chai")

class MasalaChai(BaseChai):
    def add_spices(self):
        print("Adding masala,lemon,sugar")

class ChaiShop:
    chai_cls=BaseChai

    def __init__(self):
        self.chai=self.chai_cls("Regular")

    def serve(self):
        print(f"Serving {self.chai.type} Chai in the shop")
    
class FancyShop(ChaiShop):
    chai_cls=MasalaChai
    
shop=ChaiShop()
shop.serve()
fancy=FancyShop()
fancy.serve()
fancy.chai.add_spices()
        
```

```python
#seeting Parent class attributes
class Chai:
    def __init__(self,type_,strength):
        self.type=type_
        self.strength =strength

class GingerChai(Chai):
    def __init__(self,type_,strength,spice_level):
        self.type=type_
        self.strength=strength
        self.spice_level=spice_level

class GingerChai_Extra(Chai):
    def __init__(self,type_,strength,spice_level):
        Chai.__init__(self,type_,strength)
        self.spice_level=spice_level

class GingerChai_SplExtra(Chai):
    def __init__(self,type_,strength,spice_level):
        super().__init__(type_,strength)
        self.spice_level=spice_level
ginger1=GingerChai("Ginger",100,20)
ginger2=GingerChai_Extra("GingerExtra",200,40)
ginger3=GingerChai_SplExtra("GingerSplExtra",300,50)
print(ginger1.type)
print(ginger2.type)
print(ginger3.type)
```

```python
#Method resolution order

class A:
    label="Class A:label"
class B(A):
    label="Class B:label"
class C(A):
    label="Class C:label"
class D(C,B):
    pass

cup=D()
print(cup.label)
print(D.__mro__) # the order changes based on the class extends (B,C) or (C,B)
```

```python
#Class Methods
class Chai:
    def __init__(self,type_,sweetness,size):
        self.type=type_
        self.sweetness =sweetness
        self.size=size
    
    @classmethod
    def from_dict(cls,values):
        return cls(values["type_"],values["sweetness"],values["size"])
    
    @classmethod
    def from_str(cls,values):
        type,sweetness,size=values.split('-')
        return cls(type,sweetness,size)
    

chai1=Chai("Masala","Low",100)
chai2=Chai.from_dict({"type_":"Allam",
                      "sweetness":"high",
                      "size":20})
chai3=Chai.from_str("Ilachi-medium-200")

print(chai1.__dict__)
print(chai2.__dict__)
print(chai3.__dict__)
```

```python
#property decorator

class Tealeaf:
    def __init__(self,age):
        self._age=age
    
    @property
    def age(self):
        return self._age+2
    
    @age.setter
    def age(self,age):
        if 1<=age<=5:
            self._age=age
        else:
            raise ValueError("Tea leaf should be between 1 and 5")

leaf=Tealeaf(4)
print(leaf.age)
leaf.age=6


        
```
