```python
## Objects
class Chai:
    type="special"

print(Chai.type)
Chai.is_hot=True
print(Chai.is_hot)

#objects
masala=Chai()
masala.is_hot=False

ginger=Chai()
print(f"masala is hot:{masala.is_hot}")
print(f"Ginger is {ginger.is_hot}")
```

```python
#attribute shadowing
class Chai:
    temperature=40
    is_highly_rated=True

cutting=Chai()
cutting.temperature=30
print(f"Cuttings Temp is {cutting.temperature}")
del cutting.temperature
print("After deleting the temperature from cutting obj")
print(f"Cuttings Temp is {cutting.temperature}")
print("The above is taken from the Class variable ")
```

```python
# 4self argument
class Chai():
    cups=100
    
    def describe_cup(self):
        print(self.cups)

chai1=Chai()
chai1.describe_cup()

chai2=Chai()
chai2.cups=200
print("After updating the cups value")
chai2.describe_cup()
```

```python
# we can access the describe from class
Chai.describe_cup(chai2)
Chai.describe_cup(chai1)
```

```python
## Init method
class Chai:
    def __init__(self,type_,size):
        self.type=type_
        self.size=size
    def summary(self):
        return f"{self.type} of {self.size} ml"
    
chai1=Chai("Masala",200)
print("Chai1:")
print(chai1.summary())
chai2=Chai("Ginger",100)
print("Chai2:")
print(chai2.summary())
```
