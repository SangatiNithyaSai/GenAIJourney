```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
#Voc size
voc_size=1000
```

```python
## One hot encoding
##we can pass single word or whole sentence
one_hot_repr=[one_hot(words,voc_size) for words in sent] 
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len=max([len(i.split()) for i in sent])
sent_len
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
print(embedded_docs)
```

```python
# we have padded the input senetences to get same lenth for all inputs
#lets create the model
model=Sequential()
model.add(Embedding(voc_size,output_dim=10,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)[0][0:4] ##each word into ten vectors
```
