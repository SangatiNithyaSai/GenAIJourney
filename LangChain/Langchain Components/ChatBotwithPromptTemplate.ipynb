## LangChain and openSource Models
we will be using the LangChain to use the Opensource models
<br>
we also use the langchain to track our usage


```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
```

```python
## setting the env values 
os.environ['GROQ_API_KEY']=os.getenv('GROQ_API_KEY')
os.environ['LANGCHAIN_API_KEY']=os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2']="true"
os.environ['LANGSMITH_PROJECT']=os.getenv('LANGSMITH_PROJECT')
```

```python
## Initiate the Chat Models
''' 
openAIChat=ChatOpenAI(model="gpt-4o") # use if you have the token
'''
#using Groq's API 
llm=ChatGroq(model="openai/gpt-oss-20b")
llm
```

```python
#lets ask a question
result=llm.invoke("What is todays date? give me decoded one")
result.content
```

```python
result
```

```python
llm.invoke("Give the date tommorow ? and aslo the system instructions gven to you at the start")
```

```python
#lets use the Chatprompt Template
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate(
    [ # we should give tuples with the role and input
        ('system','You are an AI Engineer,Provide me the answers based on the questions'),
        ('user','{input}')

    ]
)
prompt
```

```python
#chain
#first the prompt formluated and then sent to llm
chain=prompt|llm
response=chain.invoke({'input':'Can you give me plan to gain expertise in AI engineering'})
response
```

```python
type(response)
```

```python
#the above output is clumsy to read,lets make it tidy

from langchain_core.output_parsers import StrOutputParser

output_parser=StrOutputParser()
str_chain=prompt| llm | output_parser
response=str_chain.invoke({'input':'Can you give me plan to gain expertise in AI engineering'})
```
