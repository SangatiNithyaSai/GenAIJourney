# How to split JSON Data

Splits the Json data with control over Chunk size.
<br>
keeps nested json elements together but if needed places netween min_chunk and max_chunk size.
<br>
For string values larger than chunk size,to cap use RecursiveTextSplitter


```python
import json
import requests

json_data_json=requests.get("https://api.smith.langchain.com/openapi.json")
json_data=json_data_json.json()

json_data_str=json.dumps(json_data_json.json())
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=100)
json_chunks=json_splitter.split_json((json_data))
```

```python
json_chunks
```

```python
##Above even the len of the chunkis greater than 300 ,but it kept it together.
## Lets try the Recursive
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=50)

text_docs=text_splitter.create_documents([json_data_str])
text_docs
```
