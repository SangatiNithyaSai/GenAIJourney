## How to split by HTML header
HTMLHeaderTextSplitter - to split the HTML document 
<br>
a. Keeps related text grouped
<br>
b. preserves context-rich information encoded in documnet structure.
<br>
Can be used with other splitters in the pipeline

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string=""""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My LangChain Project Page</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            margin: 40px;
            background-color: #f4f4f4;
        }
        .container {
            background: white;
            padding: 20px;
            border-radius: 8px;
            box-shadow: 0 2px 5px rgba(0,0,0,0.1);
        }
        h1 { color: #333; }
    </style>
</head>
<body>

    <div class="container">
        <h1>Welcome to my GenAI Journey</h1>
        <p>This page is hosted from my local environment.</p>
        
        <h3>Project Status:</h3>
        <ul>
            <li>Python 3.10 Venv: <b>Active</b></li>
            <li>LangChain: <b>Installed</b></li>
            <li>Text Splitter: <b>Recursive</b></li>
        </ul>

        <button onclick="alert('Hello from the browser!')">Click Me</button>
    </div>

</body>
</html>
"""
```

```python
headers_to_split_on=[
    ("h1","Header 1"),
    ("h3","Header 3")
]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_doc_splits=html_splitter.split_text(html_string)
html_doc_splits
```

```python
## splitting from URL
url="https://www.dish.com/"

headers_to_split_on=[
##    ("title","Title"), -- we need to provide the Hierarchical 
# category like h1,h2,.. .
# For title,div,.. use the RecursiveCharacter text splitter
    ("h1","Header 1"),
    ("h2","Header2")
    
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits=html_splitter.split_text_from_url(url)
html_header_splits
```
