## HuggingFace Embeddings


```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HF_TOKEN']=os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
```

```python
text=" I am trying to learn HuggingFace Framework"
embedded_text=embeddings.embed_query(text)
embedded_text
```

```python
len(embeded_text) ## the text is embeded into 384 dimensions
# we receive the single vector array for whole text as
# it captures the overall context and the relativity semantically.
```

```python
# lets see how documnets are embedded
embedded_docs=embeddings.embed_documents([text,"I am a bit nervous"])
print(f" The first document dimensions {len(embedded_docs[0])}")
print(f" The Second document dimensions {len(embedded_docs[1])}")
```
