## Ollama 
An opensource model framework that runs on the device and can be used for the RAG applications


```python
from langchain_ollama.embeddings import OllamaEmbeddings
```

```python
embeddings=OllamaEmbeddings(model="all-minilm:latest")
```

```python
embeddings
```

```python
embed_text=embeddings.embed_documents(
    ["I am getting stronger",
    "I want to be Invincible"
    ]
)
embed_text
```
