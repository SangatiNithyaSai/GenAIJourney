### This section covers the functionalities of HuggingFace with LangChain

```python
import os 
from dotenv import load_dotenv
load_dotenv()
```

### HuggingFace Endpoint
There are two ways to use the models 
- Serverless (Free/Pro) access to limited set of models
- Dedicated Inference Endpoints- used to deploy models on Cloud


```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
repo_id="openai/gpt-oss-120b:groq"

llm=HuggingFaceEndpoint(repo_id=repo_id,temperature=0.7,huggingfacehub_api_token=os.getenv("HF_TOKEN"))
Chat_model=ChatHuggingFace(llm=llm)
```

```python
from langchain_core.messages import AIMessage,HumanMessage
messages=[
    HumanMessage("What is diffence between GenAI and Macinelearning")
]
response=Chat_model.invoke(messages)
response.content
```

```python
#the model is developed by Beijing Academy of Artificial Intelligence (BAAI)
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
model_name="BAAI/bge-small-en-v1.5"
model_kwargs={"device":"cpu"}
encode_kwargs={"normalize_embeddings":True}
hf=HuggingFaceBgeEmbeddings(
    model_name=model_name,model_kwargs=model_kwargs,encode_kwargs=encode_kwargs)

hf
```

```python
import os
from huggingface_hub import InferenceClient
os.environ["HF_TOKEN"]=os.getenv("HF_TOKEN")
client = InferenceClient(
    provider="hf-inference",
    api_key=os.environ["HF_TOKEN"]
)

result = client.feature_extraction(
    "Today is a sunny day and I will get some ice cream.",
    model="BAAI/bge-small-en",
)
```
