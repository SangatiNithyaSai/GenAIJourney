## Messages
The fundamental unit of context.Each message has the following section:
<br>
Role: Identifies the message type(user,system)
<br>
Content: represents the actual content of the message(Text,Image,Audio,documents)
<br>
Metadata-optional fields like Message ID ,Token usage
Note: The standard message type is common across all the models in Langchain

## Message Types
<br>
1. System Message - to provide instructions and context to the Model 
<br>
2. HumanMessage - to provide user input 
<br>
3. AIMessage -It contains the response from the AI 
<br>
4. Tool Message - If a model supports tool execution, we can pass the tool result back to model for context improvement


```python
from langchain_core.messages import SystemMessage,AIMessage,HumanMessage

messages=[
    SystemMessage("You are a poetry expert"),
    HumanMessage("Write a poem on LLM")
]
response=model.invoke(messages)
response.content
```

```python
system_msg=SystemMessage("You are a helpful assistant")
messages=[
    system_msg,
    HumanMessage("How do I create a Fast API")
]
response=model.invoke(messages)
response.content
```

```python
## More focused system instruction
system_msg=SystemMessage(""" 
                         You are a helpful coding assistant.You will provide the answers in a step wise manner.
                         Give example also""")
messages=[
    system_msg,
    HumanMessage("How do I create a Fast API")
]
response=model.invoke(messages)
response.content
```

```python
#Metadata
human_msg=HumanMessage(
    content="hello!",
    name="nithya",
    id="msg_123"
)
```

```python
response=model.invoke([human_msg]) # if it is a msg type, pass as a list
response.content
```

```python
ai_msg=AIMessage("I am happy to help you")
messages=[
    SystemMessage("you are a helpful assistant"),
    HumanMessage("Hi! can you help me"),
    ai_msg,
    HumanMessage("What is 2+2")
]
response=model.invoke(messages)
response.content
```

```python
response.response_metadata
```

```python
## Tool Message
from langchain_core.messages import ToolMessage
# we will pass the tool messages
ai_message=AIMessage(
    content=[],
    tool_calls=[
        {
            "name":"get_weather",
            "args":{"location":"Hyderabad"},
            "id":"call_123"
        }
    ]
)

#Lets create a tool messages
weather_result="Sunny 38 degrees"
tool_message=ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # same as the above id
)

messages=[
    HumanMessage("How is weather in Hyderabad?"),
    ai_message,
    tool_message
]
response=model.invoke(messages)
```
