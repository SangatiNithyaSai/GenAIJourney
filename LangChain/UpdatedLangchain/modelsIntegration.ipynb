## Integrating OpenAI,Groq, Gemini models
### using Batch and stream

```python
import os
from dotenv import load_dotenv
load_dotenv()
##getting api keys
os.environ['GROQ_API_KEY']=os.getenv('GROQ_API_KEY')
os.environ['GOOGLE_API_KEY']=os.getenv('GOOGLE_API_KEY')
```

```python
## the below code is for integrating OPENAI
from langchain_openai import ChatOpenAI
model=ChatOpenAI(model="gpt-4.1")
model.invoke("Hey what is the value of dollar in INR")
```

```python
##Groq 
from langchain.chat_models import init_chat_model
model=init_chat_model(model="groq:qwen/qwen3-32b")
model.invoke("Hello ! how are you")
```

```python
# other way is to directly use the ChatGroq
from langchain_groq import ChatGroq
model=ChatGroq(model="qwen/qwen3-32b")
model.invoke("How do parrots talk")
```

```python
## gemini model

model_gemini=init_chat_model(model="google_genai:gemini-2.5-flash")
model_gemini.invoke("Hey , what direction sun rises")
```

```python
#other way
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
model.invoke("tell me about yourself in 1 sentecne")
```

```python
## Streaming- to get the output as it is generated

for chunk in model.stream("Hey! tell me poem on men"):
    print(chunk.content,end="-----",flush=True)
```

```python
#Batch Processing - to send multiple requests at one go 

responses=model.batch(
    [
        "why do parrots have colourful feathers",
        "why do humans cry",
        "why is sky blue"
    ]
)
for response in responses:
    print(response)
```

```python
## batch with max concurrency
responses=model.batch(
    [
        "why do parrots have colourful feathers",
        "why do humans cry",
        "why is sky blue"
    ],
    config={
        'max_concurrency':2
    }
)
for response in responses:
    print(response)
```
