```python
import os 
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq  import ChatGroq

llm=ChatGroq(model="openai/gpt-oss-120b"
)
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate(
    [
        ("system","You are an expert in language translation,convert the input based on user requested language and give the final outpt only"),
        ("user","Convert the sentence {text} into {language}")

    ]
)
```

```python
chain=prompt| llm

response=chain.invoke({"text":"Surrender right now ",
              "language":"Telugu"})
```
